```python
# 7867
#Final prompt being used with the Python DF Agent in the Python DF Agent Strategy Notebook
offer_search_prompt = PromptTemplate(
    input_variables = ['Retailer','Brand','Category','Product','offer_retailer'],
    template = '''
    
    
    -------------------------------------------------------------
    # Role and Responsibility
    -------------------------------------------------------------
    
    - Use {offer_retailer} to return all offers that following the conditions below: 
        
    -------------------------------------------------------------
    # Conditions
    -------------------------------------------------------------
    
    - If {Retailer} is provided, pull all offers from {offer_retailer} associated with the {Retailer} in the Retailer column
    - If {Brand} is provided, pull all offers from {offer_retailer} associated with the {Brand} in the Brand column
    - If {Category} is provided, pull all offers from {offer_retailer} associated with the {Category} in the Category column
    - If {Product} is provided, pull all offers from {offer_retailer} associated with the {Product} in the Product column
    - Out of {Retailer}, {Brand}, {Category} and {Product}, choose those that have a non-null keyword value and pull 
    associated offers
    

    -------------------------------------------------------------
    # Data To Be Used
    -------------------------------------------------------------
    
    Dataset to Use:
    **********************
    {offer_retailer}
    
    Use following columns from {offer_retailer} dataset to search the keywords for associated value:
    *****************************************************************************************
    
    Value    -->      Column Name
    {Retailer} -->    {offer_retailer}.RETAILER
    {Brand} -->       {offer_retailer}.BRAND
    {Category} -->    {offer_retailer}.Category
    {Product} -->     {offer_retailer}.Product
    
    
    -------------------------------------------------------------
    # Format of Final Offer List
    -------------------------------------------------------------
    
    1. Offer 1 
    2. Offer 2
    3. Offer 3
    4. Offer 4
    5. Offer 5 ...
    
    
    List of Offers:
    ....
    
   
    ''')
```
