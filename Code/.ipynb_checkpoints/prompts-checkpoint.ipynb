```python
#Importing the relevant libraries

from langchain import PromptTemplate
```

```python
# 7867
#Final prompt being used with the Python DF Agent in the Python DF Agent Strategy Notebook
offer_search_prompt = PromptTemplate(
    input_variables = ['Retailer','Brand','Category','Product','offer_retailer'],
    template = '''
    
    
    -------------------------------------------------------------
    # Role and Responsibility
    -------------------------------------------------------------
    
    - Use {offer_retailer} to return all offers that following the conditions below: 
        
    -------------------------------------------------------------
    # Conditions
    -------------------------------------------------------------
    
    - If {Retailer} is provided, pull all offers from {offer_retailer} associated with the {Retailer} in the Retailer column
    - If {Brand} is provided, pull all offers from {offer_retailer} associated with the {Brand} in the Brand column
    - If {Category} is provided, pull all offers from {offer_retailer} associated with the {Category} in the Category column
    - If {Product} is provided, pull all offers from {offer_retailer} associated with the {Product} in the Product column
    - Out of {Retailer}, {Brand}, {Category} and {Product}, choose those that have a non-null keyword value and pull 
    associated offers
    

    -------------------------------------------------------------
    # Data To Be Used
    -------------------------------------------------------------
    
    Dataset to Use:
    **********************
    {offer_retailer}
    
    Use following columns from {offer_retailer} dataset to search the keywords for associated value:
    *****************************************************************************************
    
    Value    -->      Column Name
    {Retailer} -->    {offer_retailer}.RETAILER
    {Brand} -->       {offer_retailer}.BRAND
    {Category} -->    {offer_retailer}.Category
    {Product} -->     {offer_retailer}.Product
    
    
    -------------------------------------------------------------
    # Format of Final Offer List
    -------------------------------------------------------------
    
    1. Offer 1 
    2. Offer 2
    3. Offer 3
    4. Offer 4
    5. Offer 5 ...
    
    
    List of Offers:
    ....
    
   
    ''')
```

```python
# 7867
#Final prompt being used with the Python DF Agent in the Python DF Agent Strategy Notebook
offer_search_prompt_string_Search = PromptTemplate(
    input_variables = ['Combined_Search_String','offer_retailer'],
    template = '''
    
    -------------------------------------------------------------
    # Role and Responsibility
    -------------------------------------------------------------
    
    Your job is to return a list of offers from {offer_retailer}.
  
    -------------------------------------------------------------
    # Process
    -------------------------------------------------------------
    
    Step 1: Find keywords from {Combined_Search_String} in the "Retailer" OR "Brand" OR "Category" OR "Product" columns in 
    the {offer_retailer} dataset
    Step 2: If there are similar keywords matched, find offers associated with the specific "Retailer" OR "Brand" OR "Category"
    OR "Product"
    - Return this as a list of offers
    
    -------------------------------------------------------------
    # Data To Be Used
    -------------------------------------------------------------
    
    Dataset to Use:
    **********************
    {offer_retailer}
    
    Columns to Use:
    **********************
    1) Retailer
    2) Brand
    3) Category
    4) Product
    

    -------------------------------------------------------------
    # Important Terminology and Hierarchy
    -------------------------------------------------------------
    
    {offer_retailer}.Retailer : Retailer associated with brand of the offer
    
    {offer_retailer}.Brand: Brand associated with the offer
    
    ---->{offer_retailer}.Category : Each brand can have 1 or more categories associated with the brand
        
        ---->{offer_retailer}.Product : Each category can have 1 or more products associated with the brand   
     
     
    -------------------------------------------------------------
    # Conditions
    ------------------------------------------------------------- 
    - Make sure that keywords from {Combined_Search_String} exists in 
            a) {offer_retailer}.Retailer OR
            b) {offer_retailer}.Brand OR
            c) {offer_retailer}.Category OR
            d) {offer_retailer}.Product
    

    -------------------------------------------------------------
    # Format of Final Offer List
    -------------------------------------------------------------
    Final list of offers should be returned as:
    1. Offer 1 
    2. Offer 2
    3. Offer 3
    4. Offer 4
    5. Offer 5 ...
    
    
    List of Offers:
    ....
    
   
    ''')
```
